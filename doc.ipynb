```python
##### as_tensor #### convierte los datos en un tensor

a = np.array([[1,2,3],[4,5,6]])
b = torch.as_tensor(a)
b[0][0] = -2
print(a,b) ### cambia el tensor, pero también el numpy array

c = [1,2,3,5]
d = torch.as_tensor(c)
d[0] = 3
print(c,d) ### si el dato inicial no era un numpy array entonces no se cambia el dato inicial
```

```python
#### from_numpy ##### convierte un numpy array en un tensor

a = np.array([[1,2,3],[4,5,6]])
t_a = torch.from_numpy(a)  ### comparten memoria, la modificación de uno altera al otro
a[0][0] = 4
t_a[1][1] = 9
print(a,t_a)
```
